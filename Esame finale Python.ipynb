```python
#non ci sono valori, quindi sarÃ  sufficente fare la somma dei nuovi casi per continente senza considerare i valori nulli
```

```python
casi_per_continente = df.groupby("continent").sum("new_cases")

casi_per_continente.reset_index(inplace=True)
casi_per_continente
```

```python
#per vedere come siano distribuiti i casi per ogni nontinente, Ã¨ piÃ¹ esplicativo procedere con un grafino
import seaborn as sns
sns.barplot(data=casi_per_continente, x = "continent", y="new_cases")
```

```python
#calcoliamo la somma di tutti i casi dell'intera popolazione, ma lo faccioamo tramite i continenti, in modo da evitare eventuali errori dovuti a variabili cumulative come "Upper middle income"  oppure "World"
p = df[df["location"] == "World"]["new_cases"].sum()
p
```

```python
#creo un dizionario con all'interno i valori e le chiavi che abbiamo ottenuto
g_1 = {
    c_1: [max_c_1],
    c_2: [max_c_2],
    "Mondo": [p]
    
}

```

```python
#per poi renderlo un DataFrame della popolazione positiva 
pf = pd.DataFrame(g_1)
pf
```

```python
#definisco l'indice assegnatogli con il nome totali, in modo da poterlo poi utilizzare come colonna, invertendo gli assi e resettando l'indice in modo che sia numerico
pf.index = ["totali"]
```

```python
pf = pf.transpose()
pf = pf.reset_index()

pf
```

```python
#grafico relazione casi su totale, viene fatto in maniera scalare per poter essere piÃ¹ leggibile

sns.barplot(data=pf, x="index", y="totali")
```

```python
#Selezionare i dati relativi all'Italia nel 2022
italia_2022 = df[df["date"].dt.year==2022]
italia_2022=italia_2022.dropna(subset=["new_cases"])
italia_2022 = italia_2022[italia_2022["new_cases"] > 0]
italia_2022=italia_2022.dropna(subset=["date"])
```

```python
# mostrare con un grafico adeguato l'evoluzione del casi totali rispetto alla data.
sns.lineplot(data=italia_2022, x="date", y= "total_cases" )
```

```python
#Mostrare poi con un grafico adeguato il numero di nuovi casi rispetto alla data (filtrare i dati se necessario).
sns.lineplot(data=italia_2022, x="date", y= "new_cases" )
```

```python
#Mostrare infine un grafico che mostra l'andamento della somma cumulativa nuovi casi del 2022, e commentare se ci sono similaritÃ  con l'andamento del numero dei casi totali;
italia_2022["somma cumulativa"] = italia_2022["new_cases"].cumsum()
sns.lineplot(data=italia_2022, x="date", y="somma cumulativa")

```
